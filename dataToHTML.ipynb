```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
citiesDF = pd.read_csv('cities.csv')
citiesDF.count()
```

```python
#df = df[df.line_race != 0]
citiesDF = citiesDF[citiesDF.Humidity <= 100]
citiesDF.count()
```

```python
#>>> df.dropna()
citiesDF = citiesDF.dropna()
citiesDF.count()
```

```python
citiesDF.to_html('cities.html')
```

```python
plt.scatter(citiesDF['Lat'], citiesDF['Humidity'], marker='o', edgecolor = 'black')
# dateList = time.localtime(time.time())
ASdate = '10/12/2019'
plt.title(f'City Latitude vs. Humidity ({ASdate})') 
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")
plt.grid()
plt.savefig("LatVSHumid.png")
plt.show()
```
